## Transaction Features

```python
transactions = pd.read_csv('data/train_data.csv.zip')
```

```python
transactions.head()
```

```python
transactions['t_dat'][0]
```

```python
customers = transactions.groupby('customer_id')['price'].median().reset_index()
customers.head()
```

```python
customers2 = transactions.groupby('customer_id')['price'].median().reset_index()
customers3 = transactions.groupby('customer_id')['price'].std().reset_index()
customers4 = transactions.groupby('customer_id')['price'].min().reset_index()
customers5 = transactions.groupby('customer_id')['price'].max().reset_index()
```

```python
customers.columns = ['customer_id', 'mean_price']
customers2.columns = ['customer_id', 'median_price']
customers3.columns = ['customer_id', 'std_price']
customers4.columns = ['customer_id', 'min_price']
customers5.columns = ['customer_id', 'max_price']
```

```python
customers = customers.merge(customers2, how='left', on='customer_id')
customers = customers.merge(customers3, how='left', on='customer_id')
customers = customers.merge(customers4, how='left', on='customer_id')
customers = customers.merge(customers5, how='left', on='customer_id')
customers.head()
```

```python
transactions['month'] = transactions.apply(lambda x: x['t_dat'][5:7], axis=1)
```

```python
transactions.head()
```

```python
customer_purchases = transactions.groupby('customer_id')['price'].count().reset_index()
customer_purchases.columns = ['customer_id', 'total_purchases']
```

```python
customers = customers.merge(customer_purchases, how='left', on='customer_id')
```

```python
customer_month_purchases = transactions.groupby(['customer_id', 'month'])['price'].count().reset_index()
customer_month_purchases.head()
```

```python
customer_month_purchases = customer_month_purchases.pivot_table(index=['customer_id','month'], columns='price')
customer_month_purchases.head()
```

```python
#https://stackoverflow.com/questions/31306741/unmelt-pandas-dataframe
customer_month_purchases = customer_month_purchases.set_index(
    ['customer_id', 'month'])['price'].unstack().reset_index()
```

```python
customer_month_purchases.head()
```

```python
customer_month_purchases.columns=customer_month_purchases.columns.tolist()
customer_month_purchases.head()
```

```python
customer_month_purchases.columns = ['customer_id', 'January', 'February', 'March', 'April', 'May', 'June',
                                   'July', 'August', 'September', 'October', 'November', 'December']
customers = customers.merge(customer_month_purchases, how='left', on='customer_id')
customers.head()
```

```python
customers.fillna(0, inplace=True)
```

```python
customers_provided.head()
```

```python
customers_provided['club_member_status'].value_counts(dropna=False)
```

```python
customers_provided['active'] = 1*(customers_provided['club_member_status']=='ACTIVE')
customers_provided['precreate'] = 1*(customers_provided['club_member_status']=='PRE-CREATE')
customers_provided['left'] = 1*(customers_provided['club_member_status']=='LEFT CLUB ')

customers_provided['regularly'] = 1*(customers_provided['fashion_news_frequency']=='Regularly')
customers_provided['monthly'] = 1*(customers_provided['fashion_news_frequency']=='Monthly')
```

```python
customers_provided['<20'] = 1*(customers_provided['age']<20)
customers_provided['20-29'] = 1*((customers_provided['age']>=20) & (customers_provided['age']<30))
customers_provided['30-39'] = 1*((customers_provided['age']>=30) & (customers_provided['age']<40))
customers_provided['40-49'] = 1*((customers_provided['age']>=40) & (customers_provided['age']<40))
customers_provided['50+'] = 1*(customers_provided['age']>=50)
```

```python
customers_provided.head()
```

```python
customers_provided.columns
```

```python
customers = customers.merge(customers_provided[
    ['customer_id', 'FN', 'Active','age', 'active', 'precreate', 
     'left', 'regularly', 'monthly', '<20', '20-29', '30-39', '40-49',
     '50+']], how='left', on='customer_id')

customers.head()
```

```python
customers_channel1 = transactions[transactions['sales_channel_id']==1].groupby('customer_id')['t_dat'].count().reset_index()
customers_channel1.head()
```

```python
customers_channel2 = transactions[transactions['sales_channel_id']==2].groupby('customer_id')['t_dat'].count().reset_index()
customers_channel2.head()
```

```python
customers_channel1.columns = ['customer_id', 'channel1']
customers_channel2.columns = ['customer_id', 'channel2']
```

```python
customers = customers.merge(customers_channel1, how='left', on='customer_id')
customers = customers.merge(customers_channel2, how='left', on='customer_id')
```

```python
customers.fillna(0, inplace=True)
```

```python
customers.head()
```

```python
customers.to_csv('data/customer_features.csv', index=False)
```

```python
product_map = {}
for article_id in articles['article_id'].tolist():
    product = articles[articles['article_id']==article_id]['product_code'].values[0]
    al = articles[articles['product_code']==product]['article_id'].tolist()
    product_map[article_id] = [x for x in al if x!= article_id]

product_map
```

```python
import random 

random.choices([1, 2, 3, 4, 5], k=2)
```

```python
for article_id in product_map:
    if len(product_map[article_id])==0:
        product = articles[articles['article_id']==article_id]['product_type_no'].values[0]
        al = articles[articles['product_type_no']==product]['article_id'].tolist()
        al = [x for x in al if x!= article_id]
        if len(al)>4:
            k = 4
        else: 
            k = len(al)
        product_map[article_id] = random.choices(al, k=k)
        
```

```python
for article_id in product_map:
    if len(product_map[article_id])==0:
        print(article_id)
```

```python
oddballs = [532138001,615576001,670590001,795921001,816814001,816846001,832168001,850239001,880106001,920602001]
```

```python
for article_id in product_map:
    if len(product_map[article_id])==0:
        product_map[article_id] = random.choices(oddballs, k=k)
        
```

```python
transactions.head()
```

```python
negatives = []
for index, row in transactions.iterrows():
    for i in product_map[row['article_id']]:
        negative = {'customer_id': row['customer_id'], 'article_id': i, 'label': 0, 'month': row['month']}
        negatives.append(negative)
```
