## Transaction Features

```python
transactions = pd.read_csv('data/train_data.csv.zip')
```

```python
transactions.head()
```

```python
transactions['t_dat'][0]
```

```python
customers = transactions.groupby('customer_id')['price'].median().reset_index()
customers.head()
```

```python
customers2 = transactions.groupby('customer_id')['price'].median().reset_index()
customers3 = transactions.groupby('customer_id')['price'].std().reset_index()
customers4 = transactions.groupby('customer_id')['price'].min().reset_index()
customers5 = transactions.groupby('customer_id')['price'].max().reset_index()
```

```python
customers.columns = ['customer_id', 'mean_price']
customers2.columns = ['customer_id', 'median_price']
customers3.columns = ['customer_id', 'std_price']
customers4.columns = ['customer_id', 'min_price']
customers5.columns = ['customer_id', 'max_price']
```

```python
customers = customers.merge(customers2, how='left', on='customer_id')
customers = customers.merge(customers3, how='left', on='customer_id')
customers = customers.merge(customers4, how='left', on='customer_id')
customers = customers.merge(customers5, how='left', on='customer_id')
customers.head()
```

```python
transactions['month'] = transactions.apply(lambda x: x['t_dat'][5:7], axis=1)
```

```python
transactions.head()
```

```python
customer_purchases = transactions.groupby('customer_id')['price'].count().reset_index()
customer_purchases.columns = ['customer_id', 'total_purchases']
```

```python
customers = customers.merge(customer_purchases, how='left', on='customer_id')
```

```python
customer_month_purchases = transactions.groupby(['customer_id', 'month'])['price'].count().reset_index()
customer_month_purchases.head()
```

```python
customer_month_purchases = customer_month_purchases.pivot_table(index=['customer_id','month'], columns='price')
customer_month_purchases.head()
```

```python
#https://stackoverflow.com/questions/31306741/unmelt-pandas-dataframe
customer_month_purchases = customer_month_purchases.set_index(
    ['customer_id', 'month'])['price'].unstack().reset_index()
```

```python
customer_month_purchases.head()
```

```python
customer_month_purchases.columns=customer_month_purchases.columns.tolist()
customer_month_purchases.head()
```

```python
customer_month_purchases.columns = ['customer_id', 'January', 'February', 'March', 'April', 'May', 'June',
                                   'July', 'August', 'September', 'October', 'November', 'December']
customers = customers.merge(customer_month_purchases, how='left', on='customer_id')
customers.head()
```

```python
customers.fillna(0, inplace=True)
```

```python
# Customer chose specific color/pattern. 
# Negative for variations of same product (dedupe will remove if that's not the case)
product_map = {}
for article_id in articles['article_id'].tolist():
    product = articles[articles['article_id']==article_id]['product_code'].values[0]
    al = articles[articles['product_code']==product]['article_id'].tolist()
    product_map[article_id] = [x for x in al if x!= article_id]

# product_map
```

```python
import random 

# random.choices([1, 2, 3, 4, 5], k=2)
```

```python
# Add in other randomly selected articles of same product type
for article_id in product_map:
    product = articles[articles['article_id']==article_id]['product_type_no'].values[0]
    al = articles[articles['product_type_no']==product]['article_id'].tolist()
    al = [x for x in al if x!= article_id]
    if len(al)>4:
        k = 4
    else: 
        k = len(al)
    product_map[article_id] = product_map[article_id] + random.choices(al, k=k)
        
```

```python
for article_id in product_map:
    if len(product_map[article_id])==0:
        print(article_id)
```

```python
for article_id in oddballs:
    product_map[article_id] = random.choices(oddballs, k=4)
        
```

```python
negatives = []
for article in product_map:
    temp = pd.DataFrame({'new_article_id': product_map[article]})
    temp['article_id'] = article
    temp['label'] = 0
    negatives.append(transactions[transactions['article_id']==article][
        ['customer_id', 'month', 'article_id']].merge(temp, how='left', on='article_id'))
    if len(negatives) % 1000 == 0:
        print(len(negatives))
    
```

```python
negatives_df = pd.concat(negatives)
```
